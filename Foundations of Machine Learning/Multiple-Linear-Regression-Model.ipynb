# Building Multiple Linear Regression Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
df = pd.read_csv("Synergix_data_preprocessed_new.csv")
```

```python
df.info()
```

```python
df = df.drop(columns=['Units_sold>1000'])
df.head()
```

```python
# One-hot encode the 'Segment' column
df_encoded = pd.get_dummies(df, columns=['Segment'], drop_first=True)

# Display the first few rows of the modified dataset
df_encoded.head()
```

```python
X = df_encoded.drop('Units_sold', axis = 1)
y = df_encoded['Units_sold']
```

```python
import statsmodels.api as sm
```

```python
# Fitting the regression model using the X features
X_with_const = sm.add_constant(X)
model_reduced = sm.OLS(y, X_with_const).fit()

# Display the summary of the refitted model
model_reduced.summary()
```

# Checking the VIF Scores

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#function to calculate the VIF score
def calculate_vif(data):
    """Calculate VIF for a dataframe."""
    vif_data = pd.DataFrame()
    vif_data["Variable"] = data.columns
    vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    return vif_data

calculate_vif(X)
```

# Removing Features with VIF Scores > 10

```python
# Iteratively remove features with the highest VIF
VIF_threshold = 10
max_vif = float('inf')
while max_vif > VIF_threshold:
        # Calculate VIF
    vif_df = calculate_vif(X)
 
    #Finding the max values of vif scores
    max_vif = vif_df["VIF"].max()
    
    # If the maximum VIF is above the threshold, remove that variable
    if max_vif > VIF_threshold:
        remove = vif_df.sort_values("VIF", ascending=False).iloc[0]

        #Dropping columns from features simultaneously
        X.drop(remove["Variable"], axis=1, inplace = True)

# Display the features and their VIFs after the iterative removal
final_vif = calculate_vif(X)
final_vif.sort_values(by='VIF', ascending = False)
```

```python
# Data for regression
X.shape, y.shape
```

```python
# Add a constant for the intercept term
X_with_const = sm.add_constant(X)

# Fit the regression model using statsmodels
model = sm.OLS(y, X_with_const).fit()
```

```python
# Display the model summary
model.summary()
```

# Removing Features with P-Values > 0.05

# Run all the above cell

```python
#A quick look at the p-values
model.pvalues
```

```python
# Identify features with p-values greater than 0.05
features_to_remove = model.pvalues[model.pvalues > 0.05].index
#If constant has a p-value > 0.05, we are not removing it from our X features: {Reason}
if 'const' in features_to_remove:
    features_to_remove = features_to_remove.drop('const')
print(features_to_remove)
```

```python
# Drop those features
X = X.drop(columns=features_to_remove)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
# Convert the scaled numpy array back to a pandas DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Add constant and fit the OLS model

X_scaled_const = sm.add_constant(X_scaled_df)
```

```python
final_desc_model = sm.OLS(y, X_scaled_const).fit()
```

# Building Predictive Model Using Sklearn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Synergix_data_preprocessed_new.csv")
```

```python
df = df.drop(columns=['Units_sold>1000'])
df.head()
```

```python
# One-hot encode the 'Segment' column
df_encoded = pd.get_dummies(df, columns=['Segment'], drop_first=True)

# Display the first few rows of the modified dataset
df_encoded.head()
```

```python
X = df_encoded.drop('Units_sold', axis = 1)
y = df_encoded['Units_sold']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets (80% train, 0% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize the Linear Regression model
lr_model = LinearRegression()
```

```python
# Train the model using the training set
lr_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
# Predict target values on the test set
y_train_pred = lr_model.predict(X_train)

# Predict target values on the test set
y_pred = lr_model.predict(X_test)
```

```python
r2_train = r2_score(y_train, y_train_pred)
print('Train r2 score: ', r2_train)

r2_test = r2_score(y_test, y_pred)
print('Test r2 score: ', r2_test)
```

```python
import joblib
filename  = 'Your_model_name.joblib'
joblib.dump(lr_model, filename)
print("Model saved successfully")
```

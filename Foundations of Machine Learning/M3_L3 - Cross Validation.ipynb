```python
#reading the data
data = pd.read_csv('Synergix_data_preprocessed_new.csv')
data.head()
```

```python
#drop Units_sold column
data.drop(columns = 'Units_sold', axis = 1, inplace = True)
```

```python
#check the data
data.head()
```

```python
#Storing the ratio in a list named Rating_ratio
Rating_ratio = []
for row in data.values:
    if(row[3]+row[4] == 0):
        if(row[6]+row[7] == 0):
            #If all the ratings are zero then overall rating ratio will also be zero
            Rating_ratio.append(0.0)
        else:
            #If only the numerator(1 and 2 star) ratings are zero then add -99999 to the list temporarily which 
            #will be taken care of in the next part of the code
            Rating_ratio.append(-99999)
    else:
        Rating_ratio.append((int(row[6])+int(row[7]))/(int(row[4])+int(row[3])));
              

#replacing -99999 with the maximum rating in the list
max_rating = max(Rating_ratio)
for x in range(len(Rating_ratio)):
    if(Rating_ratio[x] == -99999):
        Rating_ratio[x] = max_rating
```

```python
#adding the column 'Good_By_Bad_Rating' to the dataframe
data['Good_By_Bad_Rating'] = Rating_ratio
```

```python
#drop the rating columns
data.drop(columns = ['1_Star_Rating', '2_Star_Rating', '3_Star_Rating', '4_Star_Rating', '5_Star_Rating'], inplace = True)
```

```python
#check the dataset
data.head().transpose()
```

```python
#get the category names
data['Segment'].unique()
```

## One-hot Encoding

```python
#one-hot encoding
data = pd.get_dummies(data, columns = ['Segment'], prefix = 'Segment', drop_first = True)
data.head()
```

```python
#Separating the independent and target variables
y = data['Units_sold>1000']
X = data.drop(columns = ['Units_sold>1000'], axis = 1)
```

## Performing KFold cross validation

<p style = 'color:green'><b>Run all the cells above before you begin</b><p>

```python
#import libraries
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.metrics import f1_score
```

```python
#define cross-validation method to use
kf = KFold(n_splits=5, random_state=42, shuffle=True)
```

```python
#setting the value k or neighbors at 10
knn = KNeighborsClassifier(10)
```

```python
#creating a pipeline when used will first scale the data and then make predictions using knn
pipeline = make_pipeline(StandardScaler(), knn)
```

Now, let us perform cross validation. The cross_validate primarily takes in these areguments- the model you want to use, the dependent and the independent variables, the folds of cross validation and the evaluation metric, and if you also want the scores on the train set.

In our case the model is part of the pipeline instance. X and y are our independent and dependent variables respectively, the cross validation folds are saved in the kf instance and the evaluation metric is the f1 score. So let us implement this code.

```python
# Perform k-fold cross-validation
cross_validate_results = cross_validate(pipeline, X, y, cv=kf, scoring='f1', return_train_score=True)
```

Lets understand what has happened when we ran this cell-
- In a loop of 5, Out of the 5 folds we created above 4 will function as training data and fifth fold as the test data
- First, the scaler.fit_transform will be applied to the training data which in our case are the 4 folds
- Then, scaler.transform will be applied on the test data which is the final fold.

After this knn model would run on the training dataset and store the f1 score of the model on the test and the train data.

And this will repeat 5 times, where in each iteration a different fold becomes the test data.

```python
test_scores = cross_validate_results['test_score']
print("Cross-validation scores:", test_scores)
print("Cross-validation Mean score:", test_scores.mean())
print("Cross-validation STD:", test_scores.std())
```

This tells us that this model has low variance. To look at the bias, we would need to look at the training scores. Lets do that in the code below.

```python
cross_validate_results['train_score']
```

The training scores are around 88%. So wecan say that the remianing 12% represents error or the information which this model was not able to capture. This is the the bias of the model. This bias is neither too high nor too low. Additionally, the gap between the train and the test score is just 3% indicating limited overfitting.

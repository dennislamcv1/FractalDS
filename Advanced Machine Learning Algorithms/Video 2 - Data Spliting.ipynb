In the next step, let's load the Synergix dataset from the current working directory. We will be using the complete version of the dataset that we retained in the previous course of this program. This version contains the SKU_ID and the dates as well.

```python
df = pd.read_csv("Synergix_data_fe_new.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df
# Split the data into 70% training and 30% remaining (test) for supervised learning
train,remaining = train_test_split(X, test_size=0.3, random_state=42)

# Split the remaining data into validation and train sets for supervised learning
# two-thirds for train and one-third for validation
val, test = train_test_split(remaining, test_size=1/3, random_state=42)

# Checking the size of each dataset to ensure the split
len_train = len(train)
len_val = len(val)
len_test = len(test)

(len_train, len_val, len_test)
```

```python
# Save the datasets to CSV files
train.to_csv("train.csv", index=False)
val.to_csv("validation.csv", index=False)
test.to_csv("test.csv", index=False)
```

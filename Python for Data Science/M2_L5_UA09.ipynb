## Solutions

```python
# Q1 - adding country to the sales data
sales.head()
```

```python
### For the above question we will be using left join.
sales=sales.merge(cust[['customer_id','country']],on='customer_id',how='left')
sales.head()
```

```python
# Q2 - country with highest profits overall
sales.groupby('country')['profit'].sum().sort_values(ascending=False)[0:1]
```

```python
# Q3 - adding vendor name to the sales data
### For the above question we will be using left join.
sales=sales.merge(vend,on='vendor_code',how='left')
sales.head()
```

```python
# Q4 - top vendor (profits) for category 'Technology'
Technology_data=sales[sales['category']=='Technology']
Technology_data.groupby('vendor')['profit'].sum().sort_values(ascending=False)[0:1]
```

```python
# Q5 - top vendor (sales) for Brazilian customers
Brazil_cust=sales[sales['country']=='Brazil']
Brazil_cust.groupby('vendor')['profit'].sum().sort_values(ascending=False)[0:1]
```

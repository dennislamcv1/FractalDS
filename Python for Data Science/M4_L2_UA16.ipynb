```python
# Q.4)
sales['segment'].unique()
```

As this is a nominal variable where order within cateories doesnt matters.

```python
from sklearn.preprocessing import OneHotEncoder
# Initialize the OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)

# Fit and transform the 'Category' column
encoded_categories = onehot_encoder.fit_transform(sales[['segment']])

# Create a DataFrame with the one-hot encoded columns
encoded_df = pd.DataFrame(encoded_categories, columns=onehot_encoder.get_feature_names_out(['segment']))

sales=pd.concat([sales,encoded_df],axis=1)
sales.head()
```

```python
# Q.5)
### In order to get customer-vendor sales details we will be merging sales data with vendor data.

sales=sales.merge(vend,on='vendor_code',how='left')
sales.head()
```

```python
sales['vendor'].nunique()
```

#### We see there are too many vendor names and if we try to convert this to a one hot encoded values , we will have 65 sparse features.
Oh we cant ingest all of them in our analysis. So what to do ?
1. One approach might be selecting the top vendor names and renaming rest into other category.
Firstly we will select what does top vendor means for us ?
If we want to include vendors who are responsible for 60% txns.

```python
sales['vendor'].value_counts(normalize=True).cumsum().loc[lambda x:x<=0.60].shape[0]
```

top 16 vendors have almost 60% of total orders. So we will keep this top 16 vendors and put rest others in 'Other' vendors catgeory.

```python
keep_vendors=list(sales['vendor'].value_counts(normalize=True).cumsum().loc[lambda x:x<=0.60].index)
keep_vendors
```

Now, with 17 categories in the data ,we can go ahead with the one-hot encoders.

**What did we achieve ?**

We preserved vendors which spanned 60% of total orders and also managed to reduced the number of categories from 64 to 16.

```python
sales['vendor_cat']=np.where(sales['vendor'].isin(keep_vendors),sales['vendor'],'Others')
sales['vendor_cat'].value_counts(normalize=True)
```

```python
# Initialize the OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)

# Fit and transform the 'Category' column
encoded_categories = onehot_encoder.fit_transform(sales[['vendor_cat']])

# Create a DataFrame with the one-hot encoded columns
encoded_df = pd.DataFrame(encoded_categories, columns=onehot_encoder.get_feature_names_out(['vendor_cat']))

sales=pd.concat([sales,encoded_df],axis=1)
sales.head()
```

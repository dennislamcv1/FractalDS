```python
# Q.3:-
from statsmodels.stats import weightstats as wst

#predefined_target
pop_sales=140

print(f'Null Hypothesis (H0): Average sales for the Office Supplies is {pop_sales}')
print(f'Alternative Hypothesis (H1): Average sales for the Office Supplies is not {pop_sales}')

alpha=0.05

#take a data where Product category is Office Supplies
df=sales[sales['category']=='Office Supplies']
```

With a larger sample size and known average population sales we will go ahead with one sample z-test.

```python
z_stat, p_value = wst.ztest(df['sales'],value=pop_sales)

print(f'Test_statistic value is : {round(z_stat,4)}, p-value is : {round(p_value,4)}')

if p_value <= alpha:
    print(f'Final Conclusion: Reject the null hypothesis')
else:
    print(f'Final Conclusion: Do not reject the null hypothesis')
```

```python
# Q.4:-
# Converting the order date into datetime foramt.
sales['order_date']=pd.to_datetime(sales['order_date'],format='%d/%m/%Y')
# Aggregating the profits on on sub-category & order date level
grouped_sales=sales.groupby(['sub_category','order_date'])['profit'].sum().unstack().T
# Filtering data for sub-categories Accessories & Art
grouped_sales=grouped_sales[['Accessories','Art']]
grouped_sales.reset_index(inplace=True)
### Filtering data for month of October
grouped_sales=grouped_sales[grouped_sales['order_date'].between('2011-10-01','2011-10-31')]
grouped_sales.sort_values('order_date',ascending=True)
print(grouped_sales.shape)
grouped_sales
```

```python
# Example paired data (before and after treatment)
Accessories = list(grouped_sales['Accessories'].dropna())
print(len(Accessories))
Art = list(grouped_sales['Art'].dropna())
print(len(Art))
```

#### Two - tailed two-sample test on the profits from *Accessories* and *Art*
- Here, we are going to test whether the profits from  Accessories and the Art are same or not
- Sample will be pciked for month of Oct 2011.
- Sample 1 is of size 20 where sub-category is Accessories.
- Sample 2 is of size 22 where sub-category is Art.
- Check whether the profits of these two samples are same

**Let us formulate the hypothesis:**
- $H_0$: Average profits of the *Accessories* and *Art* are same
- $H_1$: Average profits of the *Accessories* and *Art* are not same
- This can be written as -
    - $H_0 : \mu_1 = \mu_2$  
    - $H_1 : \mu \ne \mu_2 $

```python
import scipy.stats as stats


# Perform a paired t-test
t_statistic, p_value = stats.ttest_ind(Accessories, Art)

# Set the significance level (alpha)
alpha = 0.05



# Print the t-statistic and p-value
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Check if the p-value is less than alpha
if p_value < alpha:
    print("Reject the null hypothesis i.e Average profits of the *Accessories* and *Art* are not same")
else:
    print("Fail to Reject the null hypothesis i.e Average profits of the *Accessories* and *Art* are same")
```

```python
# Q.5:-
#create a contingency table : 
sales_filter=sales[(sales['category'].isin(['Furniture','Technology'])) & (sales['ship_mode'].isin(['Second Class','Same Day']))] 
obs=pd.crosstab(sales_filter['category'],sales_filter['ship_mode'])
obs
```

As we could see that we are supposed to check the association between 2 binary categorical variables we will go ahead with chi-square(2x2) contingency test.

```python
from scipy.stats import chi2_contingency,chisquare

print('Null Hypothesis (H0): There is no relationship between Selected product category and Selected Ship mode. (independent)')
print('Alternative Hypothesis (H1): There is a relatinoship between Selected product category and Selected Ship mode. (dependent)')


chi_sq_Stat, p_value, deg_freedom, exp_freq=chi2_contingency(obs)

print(f'Test_statistic value is : {round(chi_sq_Stat,4)}, p-value is : {round(p_value,4)}')

alpha= 0.05
if p_value < alpha:
    print(f'Final Conclusion: Reject null hypothesis i.e there exists a relationship /association between 2 variables.')
else:
    print(f'Final Conclusion: Do not reject null hypothesis.  i.e there doesnt exists a relationship /association between 2 variables.')
```
